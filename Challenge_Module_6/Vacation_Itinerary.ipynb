```python
# Import the dependencies.
import pandas as pd
import gmaps
import requests
# Import the API key.
from config import g_key
```

```python
# Store the CSV you saved created in part one into a DataFrame & read the CSV
hotel_df = pd.read_csv("Data/WeatherPy_vacation.csv")
hotel_df.head()
```

```python
hotel_df
```

```python
# To check the country column to determine the number of cities in the data 
hotel_df['Country'].value_counts()
```

```python
# Create a dataframe to obtain information about all US cities
us_vacation_df= hotel_df[hotel_df["Country"]=='US'].copy().reset_index(drop=True)
us_vacation_df
```

```python
#Drop all NaN's and reset index
us_vacation_df.dropna(inplace= True)
us_vacation_df.reset_index(drop =True, inplace = True)
us_vacation_df.sort_values(['Lat','Lng'],inplace=True)
us_vacation_df.reset_index(drop = True)
```

```python
#Create a DataFrame for the identified cities
city_index=[13,14,15,16]

vacation_itinerary = us_vacation_df.iloc[city_index].copy()
vacation_itinerary.reset_index(drop = True,inplace =True)
vacation_itinerary
```

```python
#Create a DataFrame for each city

cities_df =[]

for index in range(vacation_itinerary.shape[0]):
    cities_df.append(vacation_itinerary.iloc[index].copy())
```

```python
# Configure gmaps to use your Google API key.
gmaps.configure(api_key=g_key)
```

```python
#Create the travel map
fig =gmaps.figure()

start_point_lat_lng = (cities_df[0]['Lat'],cities_df[0]['Lng'])
end_point_lat_lng = (cities_df[3]['Lat'],cities_df[3]['Lng'])
waypoints=[(cities_df[i]['Lat'],cities_df[i]['Lng']) for i in range (1,len(cities_df)-1)]

layer_map = gmaps.directions_layer(start_point_lat_lng, end_point_lat_lng,waypoints= waypoints,travel_mode='BICYCLING')
fig.add_layer(layer_map)
fig
```

```python
# Add markers to map
fig =gmaps.figure()

locations = vacation_itinerary[["Lat", "Lng"]]

info_box_template = """
<dl>
<dt>Hotel Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
<dt>Max Temp</dt><dd>{Max Temp} Â°F</dd>
</dl>
"""

# Store the DataFrame Row.
hotel_info = [info_box_template.format(**row) for index, row in vacation_itenerary.iterrows()]



marker_layer = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(marker_layer)

# Call the figure to plot the data.
fig
```
